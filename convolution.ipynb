#Blurring


```python
import numpy as np
```

```python
import cv2
from scipy import signal
```

```python
im=img[:,:,:3]
```

```python
imgray=cv2.cvtColor(im,cv2.COLOR_RGB2GRAY)
plt.imshow(imgray)
```

To blur with more intensity increase the size of mask matrix or, blur the image again and again with the mask


```python
smoothMask=np.ones((15,15))/225
```

```python
imBlurred=signal.convolve2d(imgray,smoothMask,boundary='symm')
imBlurred1=signal.convolve2d(imBlurred,smoothMask,boundary='symm')
```

```python
fig,(ax1,ax2,ax3)=plt.subplots(1,3,figsize=(10,20))
ax1.imshow(imgray,cmap='gray')
ax2.imshow(imBlurred,cmap='gray')
ax3.imshow(imBlurred1,cmap='gray')
```

#Edge Detection

```python
xMask=np.array([[-1,0,1],[-1,0,1],[-1,0,1]])
yMask=xMask.T.copy()
```

```python
fx=signal.convolve2d(imgray,xMask,boundary='symm',mode='same')
fy=signal.convolve2d(imgray,yMask,boundary='symm',mode='same')
```

```python
fx.shape
```

```python
fy.shape
```

```python
imgray.shape
```

Gradient magnitude

```python
Gm=(fx**2+fy**2)**0.5
```

```python
plt.imshow(Gm,cmap='gray')
```

Threshold


```python
th=Gm.max()-8*Gm.std()
th
```

```python
E=Gm>th
plt.imshow(E,cmap='gray')
```
